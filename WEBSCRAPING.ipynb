```python
import asyncio
import json
from playwright.async_api import async_playwright

async def scrape_neetcode150():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        await page.goto("https://neetcode.io/practice/practice/neetcode150", wait_until="networkidle")
        await page.wait_for_selector("a[href*='/problems/']", timeout=15000)

        problems = await page.evaluate("""
            () => {
                const links = document.querySelectorAll("a[href*='/problems/']");
                return Array.from(links).map(a => ({
                    name: a.innerText.trim(),
                    href: a.href
                })).filter(p => p.name.length > 0);
            }
        """)

        await browser.close()
        return problems

problems = await scrape_neetcode150()

for p in problems:
    print(f"{p['name']} — {p['href']}")
```

```python
from playwright.async_api import async_playwright
import json

captured = []

async with async_playwright() as p:
    browser = await p.chromium.launch(headless=False)  # visible browser
    page = await browser.new_page()

    # Intercept ALL network responses
    async def handle_response(response):
        url = response.url
        if response.status == 200:
            try:
                data = await response.json()
                captured.append({"url": url, "data": data})
                print(f"✅ JSON captured: {url}")
            except:
                pass  # not JSON, skip

    page.on("response", handle_response)

    await page.goto("https://neetcode.io/practice/practice/neetcode150", wait_until="networkidle")
    await page.wait_for_timeout(5000)

    await browser.close()

# Show all captured API calls
for item in captured:
    print("\n🔗 URL:", item["url"])
    print(json.dumps(item["data"], indent=2)[:300])
```

```python
import threading
import json
from playwright.sync_api import sync_playwright

captured = []

def run_scraper():
    with sync_playwright() as p:
        browser = p.chromium.launch(headless=True)
        page = browser.new_page()

        def handle_response(response):
            try:
                if response.status == 200:
                    data = response.json()
                    captured.append({"url": response.url, "data": data})
                    print(f"✅ {response.url}")
            except:
                pass

        page.on("response", handle_response)
        page.goto("https://neetcode.io/practice/practice/neetcode150", wait_until="networkidle")
        page.wait_for_timeout(5000)
        browser.close()

# Run in a separate thread to avoid event loop conflict
t = threading.Thread(target=run_scraper)
t.start()
t.join()

print(f"\nTotal JSON responses captured: {len(captured)}")
```

```python
# Also search for arrays of slugs pattern
import re

# Look for arrays of problem slugs like ["two-sum","valid-anagram",...]
matches = re.findall(r'\[("[a-z][a-z0-9\-]+"(?:,"[a-z][a-z0-9\-]+")+)\]', r.text)
for m in matches:
    items = m.split(",")
    if len(items) > 10:  # only large arrays
        print(f"Array with {len(items)} items:")
        print(m[:300])
        print("---")
```

```python
import re
import csv
import json

# Extract all problem objects from the JS
pattern = r'\{problem:"([^"]+)",pattern:"([^"]+)",link:"([^"]+)",video:"[^"]*",difficulty:"([^"]+)",code:"[^"]*"([^}]*)\}'
matches = re.findall(pattern, r.text)

print(f"Total problems found: {len(matches)}")

problems = []
for match in matches:
    name, pattern_tag, link, difficulty, flags = match
    
    # Check if neetcode150:!0 is in the flags
    if "neetcode150:!0" in flags:
        # Get the ncLink (neetcode url slug)
        nc_link_match = re.search(r'ncLink:"([^"]+)"', flags)
        nc_slug = nc_link_match.group(1).rstrip("/") if nc_link_match else link.rstrip("/")
        
        problems.append({
            "name": name,
            "tag": pattern_tag,
            "difficulty": difficulty,
            "leetcode_url": f"https://leetcode.com/problems/{link}",
            "neetcode_url": f"https://neetcode.io/problems/{nc_slug}",
        })

print(f"NeetCode 150 problems: {len(problems)}")

# Save to CSV
with open("neetcode150.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "tag", "difficulty", "leetcode_url", "neetcode_url"])
    writer.writeheader()
    writer.writerows(problems)

print("Saved to neetcode150.csv")
for p in problems:
    print(f"[{p['difficulty']}] {p['name']} -> {p['neetcode_url']}")
```
