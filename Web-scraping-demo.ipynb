```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get("https://www.imdb.com/search/title/?moviemeter=,10", headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'html.parser')
print(soup.prettify())
```

```python
series=soup.find_all('div', class_='ipc-metadata-list-summary-item__tc')
```

```python
len(series)
```

```python
name=[]
rating=[]
desc=[]
for i in series:
    name.append(i.find('h3').text)
    rating.append(i.find('span',class_='ipc-rating-star--rating').text.strip())
    desc.append(i.find('div',class_='ipc-html-content-inner-div').text)
d={'name':name,'rating':rating,'description':desc}
df=pd.DataFrame(d)
df.style.hide(axis="index")
df
```
